# تبدیل متن فارسی به فینگلیش
در طی این پروژه تیم نرم افزار برای کار با مدل تاکاترون 2 علاوه بر متن فارسی و صوت هر یک از جملات مورد بررسی در دیتاست به فینگلیش شده آن متن فارسی نیز نیاز دارند، بنابراین یکی از مراحل و کار هایی که باید انجام میشد تبدیل متن فارسی به فینگلیش بود برای این کار در طی دو مرحله کار پیش رفت.

## تبدیل دستی
ابتدا در حدود 5000 جمله فارسی به صورت دستی و در طی تلاش سه نفر کاملا به فینگلیش تبدیل شدند

## تبدیل اتوماتیک
پس از تبدیل جملات طی پردازشی این جملات همگی به صورت دیتاستی از واژگان تبدیل شدند که شامل 38367 لغت فارسی و مطابق آنها به همین تعداد لغت  فینگلش شده وجود داشت.
